# matplotlib by plt plots

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
a=np.array([1,2,3,4,5])
b=np.array([2,4,6,8,3])
plt.title('l')
plt.xlabel('a')
plt.ylabel('b')
plt.plot(a,b)
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([2,4,6,8,3])
plt.title('l')
plt.xlabel('a')
plt.ylabel('b')
plt.plot(a,b,'--g')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([2,4,6,8,3])
plt.title('l')
plt.xlabel('a')
plt.ylabel('b')
plt.bar(a,b)
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([2,4,6,8,3])
plt.title('l')
plt.xlabel('a')
plt.ylabel('b')
plt.plot(a,b, color='r')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([2,4,6,8,3])
plt.title('l')
plt.xlabel('a')
plt.ylabel('b')
plt.hist(a)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x=['sun','mon','tue','wed','thur','fri','sat']
y=[2.3,5,6,7,3.4,2,4.8]
```

```python
ax=sns.stripplot(x,y)
ax.set(xlabel='days',ylabel='spent')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.stripplot(x='class',y='sepal length',data=dsiris)
plt.title('iris')
plt.show()
```
